Code for **super-resolution** (figures $1$ and $5$ from main paper).. Change `factor` to $8$ to reproduce images from fig. $9$ from supmat.

You can play with parameters and see how they affect the result. 

# Import libs

```python
from __future__ import print_function
import matplotlib.pyplot as plt
%matplotlib inline

import argparse
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'

import numpy as np
from models import *

import torch
import torch.optim

from skimage.measure import compare_psnr
from models.downsampler import Downsampler

from utils.sr_utils import *

torch.backends.cudnn.enabled = True
torch.backends.cudnn.benchmark =True
dtype = torch.cuda.FloatTensor

imsize = -1 
factor = 4 # 8
enforse_div32 = 'CROP' # we usually need the dimensions to be divisible by a power of two (32 in this case)
PLOT = True
```

```python
Tset = '14'
for Timg in ['010']:
    for NET_TYPE in ['skip', 'UNet', 'ResNet']:

        path_to_image = "data/testdata/super-resolution/Set{}/img_{}_SRF_2_HR.png".format(Tset, Timg)
        SAVE = False 
        if Timg == '001':
            SAVE = True

# Load image and baselines

        # Starts here
        imgs = load_LR_HR_imgs_sr(path_to_image , imsize, factor, enforse_div32)

        imgs['bicubic_np'], imgs['sharp_np'], imgs['nearest_np'] = get_baselines(imgs['LR_pil'], imgs['HR_pil'])

        if PLOT:
            plot_image_grid([imgs['HR_np'], imgs['bicubic_np'], imgs['sharp_np'], imgs['nearest_np']], 4,12);
            print ('PSNR bicubic: %.4f   PSNR nearest: %.4f' %  (
                                                compare_psnr(imgs['HR_np'], imgs['bicubic_np']), 
                                                compare_psnr(imgs['HR_np'], imgs['nearest_np'])))

# Set up parameters and net

        input_depth = 32

        INPUT =     'noise'
        pad   =     'reflection'
        OPT_OVER =  'net'
        KERNEL_TYPE='lanczos2'

        LR = 0.01
        tv_weight = 0.0

        OPTIMIZER = 'adam'

        num_iter = 3010
        reg_noise_std = 0.03

        net_input = get_noise(input_depth, INPUT, (imgs['HR_pil'].size[1], imgs['HR_pil'].size[0])).type(dtype).detach()

        net = get_net(NET_TYPE, input_depth, 3).type(dtype)

        # Losses
        mse = torch.nn.MSELoss().type(dtype)

        img_LR_var = np_to_torch(imgs['LR_np']).type(dtype)

        downsampler = Downsampler(n_planes=3, factor=factor, kernel_type=KERNEL_TYPE, phase=0.5, preserve_size=True).type(dtype)

# Define closure and optimize

        def closure():
            global i, net_input

            if reg_noise_std > 0:
                net_input = net_input_saved + (noise.normal_() * reg_noise_std)

            out_HR = net(net_input)
            out_LR = downsampler(out_HR)

            total_loss = mse(out_LR, img_LR_var) 

            if tv_weight > 0:
                total_loss += tv_weight * tv_loss(out_HR)

            total_loss.backward()

            # Log
            psnr_LR = compare_psnr(imgs['LR_np'], torch_to_np(out_LR))
            psnr_HR = compare_psnr(imgs['HR_np'], torch_to_np(out_HR))
            print ('Iteration %05d    PSNR_LR %.3f   PSNR_HR %.3f' % (i, psnr_LR, psnr_HR), '\r', end='')

            # History
            if i % 50 == 0:
                psnr_history.append([i, psnr_HR])

            if PLOT and i % 100 == 0:
                out_HR_np = torch_to_np(out_HR)
                plot_image_grid([imgs['HR_np'], imgs['bicubic_np'], np.clip(out_HR_np, 0, 1)], factor=13, nrow=3)

            if SAVE and i in progress_ckp:
                sr_img_history.append(np.clip(out_HR_np, 0, 1))

            i += 1

            return total_loss

        psnr_history = []
        progress_ckp = [0, 100, 200, 500, 1000, 2000, 3000]
        sr_img_history = []

        net_input_saved = net_input.detach().clone()
        noise = net_input.detach().clone()

        i = 0
        p = get_params(OPT_OVER, net, net_input)
        optimize(OPTIMIZER, p, closure, LR, num_iter)

        if SAVE:
            try:
                os.makedirs("data/testdata/super-resolution/Set{}/{}/".format(Tset, Timg))
            except:
                pass

            folder_path = "data/testdata/super-resolution/Set{}/{}/".format(Tset, Timg)
            save_image_grid(folder_path + NET_TYPE + '.png', sr_img_history, 7, 12)
            with open(folder_path + NET_TYPE + '.txt','a') as f:
                for itr, psnr in psnr_history:
                    f.write('%5d %f\n' % (itr, psnr))
        else:
            file_name = "data/testdata/super-resolution/Set{}/other.txt".format(Tset)
            with open(file_name,'a') as f:
                psnr = str(psnr_history[-1][-1])
                f.write('{} {} {}\n'.format(Timg, NET_TYPE, psnr))
```
